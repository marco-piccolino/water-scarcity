# Appendix C: Interviews

## *Maintaining and reviving the Foggaras water conservation and irrigation system in South Algeria* with Mohammed Mediani

**Name:** Mohammed Mediani  

**Role:** Statistical analyst and engineer during the projects:

- *Inventory, analysis and valorization of traditional water techniques of European and Saharan drainage tunnels*, funded by Euro-Mediterranean Information System on know-how in the water sector *[EMWIS, Foggaras, 2008]*

Collaborated between 2004 and 2006.

**Project location:** Southern Algeria’s Saharan oases

**Context:** The climate in Southern Algeria’s Saharan oases is extremely arid, with max 50mm of rainfall annually.  
Demographic growth, intensification of agriculture (esp. cereals), and hence degradation of water resources (level and rate of flow diminishing from one day to the next). Cultivating increasingly large plots, inhabitants have dug deep wells, introduced industrial products (such as chemical fertilizers) abandoning traditional knowledge.  
Water crisis was causing residents to leave the region; many oases had suffered desertification and invasion by sands. Ancient techniques used to catch water in catchment tunnels (*foggaras*) had been almost completely given up (*[TKWB, OASI, 2013]*)

**Project goals and activities:**
- inventory of existing drainage tunnels
- analysis of their functioning and methods of realization
- causes of persistence or abandonment
- study on restoration and maintenance methods
- potentialities for re-using and reintroducing foggaras

Mohammed was tasked with helping find ways to make the conservation and maintenance of *foggaras* sustainable.

## *Frugal technologies to help overcome water scarcity in developing countries* with Roberto Verzino

**Name:** Roberto Verzino  

**Role:** Active and long-term volunteer member of *Sermig Re.Te. per lo Sviluppo Onlus*, an Italian NGO developing frugal solutions to address various issues, including water scarcity

**Project locations:** rural communities in several countries in Africa, South America and the Near East

**Context:** Sermig Re.Te. is an Italian NGO made entirely of volunteers which is part of the wider Sermig organization. Re.Te. means "technological restitution". Focus of the organization is to help communities in several countries respond to challenges that can be addressed with the aid of frugal technical and technological innovations. The organisation is inspired by the life and works of prof. Giorgio Ceragioli *[Wiki, Ceragioli, 2022]*. Projects and requests are brought by other NGOs that operate in the target countries. Re.Te. also carries on R & D activities
